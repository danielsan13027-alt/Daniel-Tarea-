```python
import pandas as pd
from tkinter import filedialog, Tk, messagebox

def procesar_con_debug():
    root = Tk()
    root.withdraw()
    root.attributes("-topmost", True)

    ruta_entrada = filedialog.askopenfilename(title="Selecciona el Excel")
    if not ruta_entrada: return

    try:
        df = pd.read_excel(ruta_entrada)
        
        # 1. IMPRIMIR COLUMNAS PARA DEPURAR
        # Esto te dirá en la consola exactamente cómo se llaman tus columnas
        print("Columnas detectadas en tu Excel:", df.columns.tolist())

        # 2. INTENTAR DETECTAR LAS COLUMNAS AUTOMÁTICAMENTE
        # Buscamos algo que se parezca a 'Fecha' e 'Inversion'
        col_fecha = None
        col_inversion = None

        for c in df.columns:
            if 'fecha' in str(c).lower(): col_fecha = c
            if 'inver' in str(c).lower() or 'monto' in str(c).lower(): col_inversion = c

        if not col_fecha or not col_inversion:
            # Si no las encuentra solo, te avisa qué nombres vio
            messagebox.showerror("Columnas no encontradas", 
                                f"No encontré 'Fechas' o 'Inversiones'.\n\n"
                                f"Tus columnas son: {df.columns.tolist()}")
            return

        # 3. PROCESAR (Agrupar y Sumar)
        print(f"Usando: '{col_fecha}' y '{col_inversion}'")
        df_resultado = df.groupby(col_fecha, as_index=False)[col_inversion].sum()

        # 4. GUARDAR
        ruta_salida = filedialog.asksaveasfilename(
            defaultextension=".xlsx",
            initialfile="Inversiones_Consolidadas.xlsx"
        )

        if ruta_salida:
            df_resultado.to_excel(ruta_salida, index=False)
            messagebox.showinfo("¡Éxito!", "Archivo procesado y guardado.")

    except Exception as e:
        print(f"Error detallado: {e}")
        messagebox.showerror("Error", f"Sucedió algo: {e}")

if __name__ == "__main__":
    procesar_con_debug()
```
