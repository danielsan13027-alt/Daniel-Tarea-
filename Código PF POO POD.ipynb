#1. Programación Funcional 


```python
# pf_local.py - Ejemplo de "Hola Mundo" en Programación Funcional

def generar_saludo(nombre: str) -> str:
    """
    Función PURA que genera el saludo.
    No depende del estado global ni modifica variables fuera de su ámbito.
    Input (nombre) -> Output (saludo). No hay efectos secundarios.
    """
    return f"¡Hola, {nombre}! Este es un saludo funcional."

def ejecutar_saludos(lista_nombres: list[str]):
    """
    Ejemplo de cómo se procesa una lista utilizando funciones puras 
    (en lugar de ciclos que modifican estado). Usamos map o list comprehension.
    """
    # Aplicamos la función 'generar_saludo' a cada elemento de la lista.
    saludos = [generar_saludo(nombre) for nombre in lista_nombres]
    return saludos

def main():
    print("="*50)
    print("          PROGRAMACIÓN FUNCIONAL (PF)")
    print("="*50)
    
    # 1. Ejemplo de llamada directa a la función pura:
    nombre_usuario = "Mundo Funcional"
    saludo_directo = generar_saludo(nombre_usuario)
    print(f"[1] Saludo simple (Función Pura): {saludo_directo}")

    # 2. Ejemplo de procesamiento de datos (manejo sin estado/mutación):
    nombres = ["Alan", "Grace", "Hopper"]
    resultados = ejecutar_saludos(nombres)
    print("\n[2] Procesando múltiples saludos:")
    for resultado in resultados:
        print(f"  -> {resultado}")

if __name__ == "__main__":
    main()
```

#2. Programación Orientada a Objetos

```python
# poo_local.py - Ejemplo de "Hola Mundo" en Programación Orientada a Objetos (POO)

class SaludoGreeter:
    """
    Representa un objeto con la capacidad (comportamiento) de generar saludos.
    Los datos y el comportamiento están encapsulados juntos en esta Clase.
    """
    def __init__(self, nombre_inicial: str):
        # 'nombre_inicial' es un atributo o dato que pertenece al estado del objeto.
        self._nombre = nombre_inicial 
        print(f"--- Objeto SaludoGreeter creado para '{self._nombre}' ---")

    def saludar(self) -> str:
        """
        Método (comportamiento) que utiliza los atributos internos del objeto.
        Este método *actúa* sobre el estado interno ('self').
        """
        return f"¡Hola, {self._nombre}! Soy un saludo encapsulado en POO."

    def cambiar_nombre(self, nuevo_nombre: str):
        """Método que cambia el estado (mutabilidad) del objeto."""
        print(f"\n[Acción de Estado] Cambiando nombre de '{self._nombre}' a '{nuevo_nombre}'.")
        self._nombre = nuevo_nombre


def main():
    print("="*50)
    print("          PROGRAMACIÓN ORIENTADA A OBJETOS (POO)")
    print("="*50)

    # 1. Creación de la instancia del Objeto (Instanciación).
    greeter = SaludoGreeter("Mundo Orientado a Objetos")
    
    # 2. Llamada al método del objeto (Accediendo al comportamiento).
    saludo_poo = greeter.saludar()
    print(f"[Resultado] {saludo_poo}")

    # 3. Demostración de mutabilidad y cambio de estado:
    greeter.cambiar_nombre("Pythonista Experto")
    nuevo_saludo = greeter.saludar()
    print(f"[Resultado] {nuevo_saludo}")


if __name__ == "__main__":
    main()
```

 3. Programación Orientada a Datos 

```python
# pod_local.py - Ejemplo de "Hola Mundo" en Programación Orientada a Datos (POD)

def procesar_saludos_por_datos(lista_nombres: list[str]) -> list[str]:
    """
    Esta función es la 'lógica' o el algoritmo puro. 
    Recibe una colección de datos y devuelve otra colección procesada, 
    sin depender del estado interno de ningún objeto.
    """
    saludos_procesados = []
    for nombre in lista_nombres:
        # La lógica se aplica sobre los datos (nombre), no sobre un 'objeto' con ese dato.
        mensaje = f"Datos procesados para: ¡Hola, {nombre}!"
        saludos_procesados.append(mensaje)
    return saludos_procesados

def main():
    print("="*50)
    print("          PROGRAMACIÓN ORIENTADA A DATOS (POD)")
    print("="*50)
    
    # 1. Los datos están separados y agrupados en una sola colección (Array/Vector).
    datos_nombres = ["Carlos", "Diana", "Eduardo"]

    print(f"[Datos Fuente] Nombres a procesar: {datos_nombres}")

    # 2. La función opera directamente sobre la colección de datos, sin necesidad
    # de crear un objeto 'Usuario' por cada nombre.
    saludos = procesar_saludos_por_datos(datos_nombres)
    
    print("\n[Resultados Procesados]:")
    for saludo in saludos:
        print(f"  -> {saludo}")

if __name__ == "__main__":
    main()
```
